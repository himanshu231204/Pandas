```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True: 
            data = data + i 
        if i == '(':
            check = True 
    
    
```

```python
df["Episodes"]=df['Title'].apply(extract_episodes) # make new columns for Epidodes
```

```python
df['Episodes']= df['Episodes'].str.replace(" eps","") # remove "eps" from data 
```

```python
df['Episodes'] = df['Episodes'].astype(int)  # convert to int type
```

```python
def extraction_time(txt): 
    check=False 
    data=" "
    for i in range(len(txt)): 
        if txt[i] ==")":
            for j in range(i+1,i+20): 
                data= data+txt[j] 

            return data
            
```

```python
df['Total Time'] = df['Title'].apply(extraction_time)
```

```python
# finding the total months
```

**total duration in months nikalna chahte ho, to hum pandas ka to_datetime aur date_diff use karke easily calculate kar sakte hain.**



```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        # Remove extra spaces
        period = period.strip()
        start_str, end_str = [s.strip() for s in period.split(' - ')]
        
        # Convert to datetime
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        
        # Calculate difference
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    
    except Exception as e:
        print(f"Error parsing: '{period}' → {e}")
        return None
```

```python
df['Months'] = df['Total Time'].apply(calculate_total_months)
```
